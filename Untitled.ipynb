```python
re_tokeniser = re.compile(r'([a-z]+)')
```

```python
def get_similarity(text_1, text_2):
    toks_1 = set(re_tokeniser.findall(text_1))
    toks_2 = set(re_tokeniser.findall(text_2))
    return len(toks_1.intersection(toks_2)) / len(toks_1.union(toks_2))
```

```python
data["y_pred"] = data[["text_1", "text_2"]].apply(lambda x : get_similarity(x.text_1, x.text_2), axis=1)
```

```python
data[["y_pred"]].to_csv("output/")
```

```python
all_questions = list(sorted(set(data.text_1).union(set(data.text_2))))
```

```python
question_text_to_idx = dict([b, a] for a, b in enumerate(all_questions))
```

```python
import harmony
from harmony.schemas.requests.text import Instrument, Question
questions = []
for idx, question_text in enumerate(all_questions):
    questions.append(Question(question_text=question_text, question_no = f"{idx}"))
instrument = Instrument(questions=questions)
```

```python
questions, similarity, query_similarity, new_vectors_dict = harmony.match_instruments([instrument])
```

```python
preds = [0] * len(data)
for idx in range(len(data)):
    text_1 = data.text_1.iloc[idx]
    text_2 = data.text_2.iloc[idx]
    idx_1 = question_text_to_idx[text_1]
    idx_2 = question_text_to_idx[text_2]
    preds[idx] = np.abs(similarity[idx_1, idx_2])
```

```python
data["y_pred"] = preds
```

```python
actual = data.y
predict = data.y_pred

corr_matrix = numpy.corrcoef(actual, predict)
corr = corr_matrix[0,1]
print(f"Correlation = {corr:.4f}")
R_sq = corr**2

print(f"RÂ² = {R_sq:.4f}")
```

```python
fpr, tpr, threshold = metrics.roc_curve(data.y, data.y_pred)
roc_auc = metrics.auc(fpr, tpr)
```
